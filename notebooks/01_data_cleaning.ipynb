```python
import numpy as np
import pandas as pd
```

```python
# read files using relative path 
df = pd.read_csv('..//data//raw//big_startup_sucess_dataset.csv')
```

```python
# check rows and columns and information about dataset
print(df.shape)
print(df.info())
```

```python
df.head()
```

```python
# convert object into datetime
df['founded_at'] = pd.to_datetime(df['founded_at'], errors = 'coerce')
df['first_funding_at'] = pd.to_datetime(df['first_funding_at'], errors='coerce')
df['last_funding_at'] = pd.to_datetime(df['last_funding_at'], errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
# year extract 
df['last_funding_year'] = df['last_funding_at'].dt.year
df['last_funding_year']
```

```python
# condition comapnies between 1995 to 2026
condition = (df['last_funding_year']>=1995) & (df['last_funding_year']<=2026)
df_cleaned = df[condition]
df_cleaned
```

```python
print("Original Data Rows:", df.shape[0])
print("Cleaned Data Rows:", df_cleaned.shape[0])
print("Total Outliers Deleted:", df.shape[0] - df_cleaned.shape[0])
```

```python
df_cleaned = df_cleaned.copy() # .copy() lagana zaroori hai
```

```python
df_cleaned['is_failed'] = np.where(df_cleaned['status'] == 'closed', 1, 0) # np.where lambda se 10x fast hai
df_cleaned
```

```python
# 1. Convert string/hyphen data into numeric values, coercing invalid strings into NaN
df_cleaned['funding_total_usd'] = pd.to_numeric(df_cleaned['funding_total_usd'], errors='coerce')

# 2. Verify and count the total number of NaN (missing) values generated after conversion
print("NaN values after conversion:", df_cleaned['funding_total_usd'].isnull().sum())

# 3. Impute all missing/NaN values in the funding column with 0 (representing no funding received)
df_cleaned['funding_total_usd'] = df_cleaned['funding_total_usd'].fillna(0)
```

```python
df_cleaned
```

```python
# 4. Final Target variable distribution & Export
print("\nTarget Distribution (is_failed):")
print(df_cleaned['is_failed'].value_counts())
```

```python
df_cleaned.to_csv('../data/processed/cleaned_startup_data.csv', index=False)
print("\nFile successfully cleaned and exported!")
```
