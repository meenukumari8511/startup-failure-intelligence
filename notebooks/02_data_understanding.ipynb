```python
import numpy as np
import pandas as pd
```

```python
# Load the Dataset
df = pd.read_csv(r"../data/raw/big_startup_success_dataset.csv")
```

```python
# Dataset Information
df.info()
```

```python
# Convert Date Columns
date_columns = [
    "founded_at",
    "first_funding_at",
    "last_funding_at"
]

for column in date_columns:
    df[column] = pd.to_datetime(df[column],errors='coerce')
```

```python
# Verify Datatypes
df.info()
```

```python
# Extract Year from Date Columns
df["founded_year"] = df["founded_at"].dt.year
df["first_funding_year"] = df["first_funding_at"].dt.year
df["last_funding_year"] = df["last_funding_at"].dt.year
```

```python
# Verify Extracted Years
df[
    [
        "founded_at",
        "founded_year",
        "first_funding_at",
        "first_funding_year",
        "last_funding_at",
        "last_funding_year"
    ]
].head()
```

```python
# Year Statistics
print("Minimum Founded Year :", df["founded_year"].min())
print("Maximum Founded Year :", df["founded_year"].max())
```

```python
# Explore Year Distribution
# Oldest Years
df["founded_year"].value_counts().sort_index().head(20)
```

```python
# Latest Years
df["founded_year"].value_counts().sort_index().tail(20)
```

```python
# Year Comparison
years = [2000, 2005, 2010, 2015]

for year in years:
    total = (df["founded_year"] >= year).sum()
    print(f"{year}+ Startups : {total}")
```

```python
# Create Years Datasets
year_datasets = {}
for year in years:
    year_datasets[year] = ( df[df["founded_year"] >= year].copy())
```

```python
# Compare Dataset Shapes
print("Original Dataset :", df.shape)

for year, subset in year_datasets.items():
    print(f"{year}+ Dataset :", subset.shape)
```

# Day-02 Summary

Today I focused on understanding the date-related information present in the startup dataset.

I converted all date columns into datetime format using `pd.to_datetime()` and safely handled invalid values using `errors="coerce"`.

After conversion, I extracted the year from each date column because yearly information is more useful for trend analysis than complete dates.

Instead of selecting a working dataset based on assumptions, I created multiple year-based subsets (2000+, 2005+, 2010+, and 2015+) for comparison.

The final working subset will be selected only after comparing data quality and business relevance in the next stage.

The objective of Day-02 was to understand the temporal structure of the dataset before performing any cleaning or filtering.


## Decision Log

### Decision-01

Converted all date columns into datetime.

Reason:

Date calculations are not possible on string values.

-----------------------------------

### Decision-02

Extracted year from every date column.

Reason:

Year-level analysis is more meaningful for startup trends.

### Decision-03
