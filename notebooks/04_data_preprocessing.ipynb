```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"../data/raw/big_startup_success_dataset.csv")
```

```python
date_columns = [
    "founded_at",
    "first_funding_at",
    "last_funding_at"
]

for column in date_columns:
    df[column] = pd.to_datetime(df[column],errors="coerce")
```

```python
# Create Year Columns
df["founded_year"] = df["founded_at"].dt.year
df["first_funding_year"] = df["first_funding_at"].dt.year
df["last_funding_year"] = df["last_funding_at"].dt.year
```

```python
# Check Current Datatype
df["funding_total_usd"].dtype
```

```python
# Replace Hidden Missing Values
df["funding_total_usd"] = df["funding_total_usd"].replace("-", np.nan)
```

```python
# Verify Missing Values
df["funding_total_usd"].isnull().sum()
```

```python
# Convert Funding Column into Numeric
df["funding_total_usd"] = pd.to_numeric(df["funding_total_usd"],errors="coerce")
```

```python
df["funding_total_usd"].dtype
```

```python
# Drop Non-Predictive Columns
columns_to_drop = ["homepage_url"]

df = df.drop(columns=columns_to_drop)
```

```python
# Final Dataset Information
df.info()
```

```python
# Create Processed Dataset
df.to_csv("../data/processed/startup_phase1.csv",index=False)
```

Today I performed the first preprocessing stage of the startup dataset.
